```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph 
```

```python
class MyState(TypedDict):
    message: str 


def process_node(state:MyState) -> MyState:
    """Create a node that processes the message in the state."""
    state["message"] = "Hello my name is " + state["message"] + ", nice to meet you!" 
    return state
```

```python
graph = StateGraph(MyState)
graph.add_node("Beo Ngu",process_node)
graph.set_entry_point("Beo Ngu")
graph.set_finish_point("Beo Ngu")

app = graph.compile()
```

```python
from IPython.display import Image, display 
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message": "Trần Thị Nhật Minh"})
result["message"]
```
